```python
countries = dict.fromkeys(names)
```

```python
for i in range(24):
    countries[names[i]] = list(datasets[i]["Area"].drop_duplicates().values)
```

```python
all_countries = sorted([*set(np.concatenate([countries[names[i]] for i in range(24)]))])
country_dict = dict.fromkeys(all_countries)
```

```python
grouped = [sets.groupby("Area") for sets in datasets]
```

```python
for country in all_countries:
    country_dict[country] = []
for group in grouped:
    for country in all_countries:
        try:
            country_dict[country].append(group.get_group(country).drop(columns="Area"))
        except:
            pass
```

## Final groups: Grouped by Country, all vars

```python
final_groups = dict.fromkeys(all_countries)
for country in all_countries:
    final_groups[country] = country_dict[country][0].append([country_dict[country][i] for i in range(1,len(country_dict[country]))],ignore_index=True)
    final_groups[country].drop(columns=["Area Id"],inplace=True)
```

## Trim groups: Custom shortened variable list
## Sort groups: Sorted by custom variable list

```python
trim_groups = dict.fromkeys(all_countries)
sort_groups = dict.fromkeys(all_countries)
for country in all_countries:
    trim_groups[country] = final_groups[country][final_groups[country][VI].isin(varids)]
    df_list = []
    for i in varids:
        df_list.append(trim_groups[country][trim_groups[country][VI]==i])
    sort_groups[country] = pd.DataFrame(pd.concat(df_list,ignore_index=True))
```

## Clean groups: NaN values removed from values column

```python
clean_groups = dict.fromkeys(all_countries)
for country in all_countries:
    clean_groups[country]=sort_groups[country].dropna(subset=['Value'])
```

## Put back into csv files

```python
for country in all_countries:
    clean_groups[country].to_csv(f"Data/by-country/{country}.csv")
```
